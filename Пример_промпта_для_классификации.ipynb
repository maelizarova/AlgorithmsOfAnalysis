**Контекст**  
Для анализа обращения — необходимо отделять саму причину жалобы (то, что спровоцировало её возникновение) от её последствий. Причина — это конкретный факт, ошибочная операция или некорректное действие системы (например, неверный расчёт комиссии, ошибка в тарифах, неправильное информирование и т.п.). Последствия (списание средств, блокировка карты, ухудшение сервиса) являются лишь результатом этой причины и не подлежат классификации. При категоризации обращения мы фиксируем именно ту проблему, из‑за которой клиент получил недовольство, т. е. классифицируем её с точки зрения клиента, а не с точки зрения внутренних процессов банка.
При формировании ответа используйте именно эти названия категорий в поле **issues**.

**Запросы клиента (requested_actions)**  
При обработке обращения необходимо выделять каждый конкретный запрос клиента – это чётко сформулированное требование выполнить определённое действие (например, «вернуть комиссию», «открыть новый счёт», «разблокировать карту», «выдать выписку» и т.п.) – и оформлять его в массив requested_actions, где указываются три обязательных поля: category (короткое название действия, используем только указанные в инструкции названия, например «Возврат комиссии», «Разблокировка карты», «Выдача выписки»), sub‑category (при необходимости уточняющая подкатегория) и description (развёрнутое описание запроса с указанием всех деталей, приведённых клиентом).
```python
def safe_load(s):
    try:
        return json.loads(s)
    except json.JSONDecodeError as e:
        # Можно логировать: print(f'Bad JSON: {e} – row {idx}')
        return None
```

```python
def get_values(dicts, key):
    if isinstance(dicts, list):
        return [d.get(key) for d in dicts]
```

```python
q=a
```

```python
q['issue'] = q.issues.apply(lambda x: get_values(x, 'category') if isinstance(x, list) else x)
q['sub_issue'] = q.issues.apply(lambda x: get_values(x, 'sub-category') if isinstance(x, list) else x)
q['issue_desc'] = q.issues.apply(lambda x: get_values(x, 'description') if isinstance(x, list) else x)
q['refusion'] = q.requested_actions.apply(lambda x: get_values(x, 'category') if isinstance(x, list) else x)
q['sub_refusion'] = q.requested_actions.apply(lambda x: get_values(x, 'sub-category') if isinstance(x, list) else x)
q['refusion_desc'] = q.requested_actions.apply(lambda x: get_values(x, 'description') if isinstance(x, list) else x)
```
