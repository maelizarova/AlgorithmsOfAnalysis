````python
import langchain
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
````

````python
LLM_KWARGS = {
    # 'top_p':0.95, 
    # 'top_k':20, 
    'max_tokens':8192,
    # 'temperature':0.1
}
BASE_URL = 'https://gpt.run.fc.uralsibbank.ru/v1'
API_KEY = 'EMPTY'
MODEL = 'openai/gpt-oss-120b'

llm = ChatOpenAI(model=MODEL, 
                 base_url=BASE_URL,
                 api_key=API_KEY, 
                 model_kwargs=LLM_KWARGS,)
````

````python
SYSTEM = ''' 
Ты — эксперт по качественной проверке классификации клиентских обращений банка.  
Твоя задача — определить, соответствует ли выбранная оператором (или автоматической) категория реальному содержанию отзыва/обращения клиента.

**Ввод:**  
1. Текст обращения клиента (полный, без сокращений).  
2. Категория, назначенная для этого обращения (одна из перечисленных ниже).

**Список допустимых категорий (выбирай только одну):**  
- Проблема с перевыпуском карты  
- Проблема с 3‑D Secure / не приходят SMS / Push  
- Платеж / перевод не исполнен / исполнен несвоевременно  
- Отказ/ошибка в проведении операций по карте  
- Не согласен с лимитом на снятие/перевод  
- Отмена комиссии за снятие/внесение в стороннем АТМ  
- Отмена комиссии за оплату ЖКУ  
- Отмена ежегодной комиссии за РКО  
- Отмена ежемесячной комиссии за РКО  
- Отмена разовой комиссии за РКО  
- Разблокировка средств по операции  
- Несвоевременное зачисление денежных средств  
- Несогласие с образованием технической / дебиторской задолженности  
- Не согласен с конвертацией / курсовой разницей  
- Клиент не согласен с удержанием НДФЛ  
- Dragon Pass  
- Не согласен с начисленными процентами  
- Mir Pass  

**Требуемый вывод:**  
Верни ответ строго в формате JSON с единственным корневым полем `Decision` — булевым значением:  

- `true` — если содержание обращения полностью совпадает с назначенной категорией (классификация правильна).  
- `false` — если содержание относится к другой теме или категоризация неверна.

**Формат ответа:**  

```json
{{
  "decision": true|false
}}

Отзыв:
{feedback}
Категория:
{category}
'''
````

````python
from pydantic import BaseModel

class Answer(BaseModel):
    decision: bool

llm_json = llm.with_structured_output(Answer)
````

````python
prompt = ChatPromptTemplate.from_template(SYSTEM)
chain = prompt | llm_json
````

````python
ix = 6
feedback = oldres.text[ix]
label = oldres.answer[ix]
````

````python
_ = chain.invoke({'feedback':feedback, 'category':label})
````

````python
_.model_dump()
````

````python
from collections import defaultdict
import numpy as np
import tqdm
# Превращаем в массив (по строкам)
arr = oldres.to_numpy()                   # shape = (4000, n_features)

batch_size = 100
n_batches  = int(np.ceil(len(arr) / batch_size))

# Делим массив на n_batches почти одинаковых кусочка
chunks = np.array_split(arr, n_batches)   # возвращает список из n_batches массивов
````

````python
!mkdir result_old_llm_as_judge
````

````python
%%time
out = defaultdict(list)
for enum, chunk in enumerate(tqdm.tqdm(chunks[:]), start=0):
    temp_df = pd.DataFrame(chunk, columns=oldres.columns)
    inputs = [{'category':label, 'feedback':feedback} for label, feedback in zip(temp_df.answer, temp_df.text)]
    c = chain.batch(inputs, config={'max_concurrency': 10})
    numbers = temp_df.number.tolist()
    answers = [c1.model_dump()['decision']  for c1 in c]
    
    out['number'].extend(numbers)
    out['answer'].extend(answers)
    a = pd.DataFrame().from_dict(out)
    # temp_df = temp_df[['number', 'product', 'type', 'sub-type','text']]
    # temp_df = temp_df.merge(a, left_on=['number'], right_on=['number'], how='left')
    a.number = a.number.astype(str)
    a.to_parquet(f'./result_old_llm_as_judge/{enum}.parquet')
    out = defaultdict(list)
    # break
    # break
````

````python
%%time
out = defaultdict(list)
for enum, chunk in enumerate(tqdm.tqdm(chunks[:]), start=0):
    temp_df = pd.DataFrame(chunk, columns=issues.columns)
    inputs = [{'feedback': feedback, 'category':label, 'sub_category':sublabel} for label, feedback, sublabel in zip(temp_df.category, temp_df.text, temp_df['sub_category'])]
    c = chain.batch(inputs, config={'max_concurrency': 5})
    numbers = temp_df.number.tolist()
    answers = [c1.model_dump()  for c1 in c]
    answers1 = [c1['category']  for c1 in answers]
    answers2 = [c1['sub_category']  for c1 in answers]
    out['number'].extend(numbers)
    out['category'].extend(answers1)
    out['sub_category'].extend(answers2)
    # out['answer'].extend(answers)
    a = pd.DataFrame().from_dict(out)
    # # temp_df = temp_df[['number', 'product', 'type', 'sub-type','text']]
    # # temp_df = temp_df.merge(a, left_on=['number'], right_on=['number'], how='left')
    a.number = a.number.astype(str)
    a.to_parquet(f'./result_new_llm_as_judge/{enum}.parquet')
    out = defaultdict(list)
    # break
    # break
````
